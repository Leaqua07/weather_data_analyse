## Aufbau

Dieses Projekt befasst sich mit der Auswertung von Wetterdaten, die mit Hilfe von Kafka über die Api der Deutschen Wetterdienstes (DWD) abgerufen werden. Es setzt sich aus folgenden Komponenten zusammen:

* Das Jupyter Notebook **01_kafka-consumer.ipynb** importiert die Daten von Kafka in die MongoDB (bearbeitet von Benedikt Funke).
* Im Jupyter Notebook **02_kafka-producer.ipynb** werden die Wetterdaten an Kafka gesendet (bearbeitet von Benedikt Funke)
* Das Jupyter Notebook **03_station_import.ipynb** befasst sich mit dem inport der Wetterstationsdaten in die MongoDB (bearbeitet von Benedikt Funke)
* Im Jupyter Notebook **04_analyse_1.ipynb** werden die Wetterstationen auf einer Karte anezeigt, von denen die verfügbaren Wetterdaten abelesen werden können (bearbeitet von Luisa Fuss).
* Das Jupyter Notebook **05_analyse_2.ipynb** zeigt den Temperaturverlauf einer ausgewählten Wetterstation an einem ausgewählten Datum aus (bearbeitet von Lea Pellikan).
